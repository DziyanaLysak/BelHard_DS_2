```python
import matplotlib.pyplot as plt
import re

# Укажи путь к loss_log.txt (твой путь)
log_path = r"C:\Users\ASUS VivoBook\PycharmProjects\BelHard_DS_2\pytorch-CycleGAN-and-pix2pix\checkpoints\night2day_100epochs\loss_log.txt"

epochs = []
g_losses = []
d_losses = []

with open(log_path, 'r') as f:
    for line in f:
        match = re.search(r'epoch:\s*(\d+).*G_GAN:\s*([\d\.]+).*D_real:\s*([\d\.]+)', line)
        if match:
            epochs.append(int(match.group(1)))
            g_losses.append(float(match.group(2)))
            d_losses.append(float(match.group(3)))

# Графики
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs, g_losses, label='Generator Loss (G_GAN)', color='blue')
plt.xlabel('Эпоха')
plt.ylabel('Loss')
plt.title('Generator Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs, d_losses, label='Discriminator Loss (D_real)', color='red')
plt.xlabel('Эпоха')
plt.ylabel('Loss')
plt.title('Discriminator Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
